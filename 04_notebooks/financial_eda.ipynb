```python
import pandas as pd
import glob
from ydata_profiling import ProfileReport

# Get a list of all Excel files that match the pattern 'combined_part_*.xlsx'
excel_files = glob.glob('/content/combined_part_*.xlsx')

# Initialize an empty list to store dataframes
df_list = []

# Loop through each Excel file, read it, and append to the list
for file in excel_files:
    try:
        df = pd.read_excel(file)
        df_list.append(df)
    except Exception as e:
        print(f"Error reading {file}: {e}")

# Concatenate all dataframes into a single dataframe
if df_list:
    combined_df = pd.concat(df_list, ignore_index=True)
    print(f"Successfully combined {len(df_list)} Excel files into a single DataFrame with {len(combined_df)} rows and {len(combined_df.columns)} columns.")
    # Display the first few rows of the combined DataFrame
    print(combined_df.head())
else:
    print("No Excel files were successfully read or found.")
    combined_df = pd.DataFrame() # Create an empty DataFrame if no files were read
```

```python
from IPython.display import display

if not combined_df.empty:
    # Generate the Profile Report
    print("Generating ydata-profiling report...")
    profile = ProfileReport(combined_df, title="Combined Excel Data Profiling Report", explorative=True)

    # Display the report in the notebook
    display(profile)
    print("Report generation complete.")
else:
    print("Cannot generate report: combined_df is empty.")
```

```python
if not combined_df.empty:
    report_filename = "combined_data_report.html"
    profile.to_file(report_filename)
    print(f"Profile report saved to {report_filename}. You can download it from the file browser.")
else:
    print("No report to save: combined_df is empty.")
```
